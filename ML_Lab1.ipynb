```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("C:/Users/Disha Devatha/Downloads/GymDataset.csv")
df.head()
```

```python
#To find missing values in each column
df.isnull().sum()
```

```python
if 'Height (m)' in df.columns:
    df['Height (m)'] = df['Height (m)'].fillna(df['Height (m)'].mode()[0])
```

```python
df.isnull().sum()
```

```python
# Bar Chart on Frequency of Weapon Used Across the Country
categorical_col = 'Height (m)'
print(df[categorical_col].value_counts())
df[categorical_col].value_counts().plot(kind='bar', color='skyblue', title='Height of people that go to the gym')
plt.show()

#Interpretation:
#This Bar Chart shows the frequency of weapons which are used to commit crimes across various parts of the country
```

```python
# Cross-tabulation and Heatmap
cat_col1, cat_col2 = 'Calories_Burned', 'Weight (kg)'
crosstab = pd.crosstab(df[cat_col1], df[cat_col2])
print(crosstab)
sns.heatmap(crosstab, annot=True, cmap='coolwarm', fmt='d')
plt.title(f'Heatmap of {cat_col1} vs {cat_col2}')
plt.show()
```

```python
# Cross-tabulation and Heatmap
cat_col1, cat_col2 = 'Water_Intake (liters)', 'Workout_Type'
crosstab = pd.crosstab(df[cat_col1], df[cat_col2])
print(crosstab)
sns.heatmap(crosstab, annot=True, cmap='coolwarm', fmt='d')
plt.title(f'Heatmap of {cat_col1} vs {cat_col2}')
plt.show()
```

```python
# Numerical vs Categorical: Boxplot
sns.boxplot(x=df['Workout_Type'], y=df['Fat_Percentage'])
plt.title(f'Boxplot of Workout Type grouped by Fat Percentage')
plt.show()
```

```python
# Scatterplot matrix of numerical variables
numerical_cols = ['BMI', 'Gender', 'Workout_Type']
sns.pairplot(df[numerical_cols], diag_kind='auto')
plt.suptitle('Scatterplot Matrix', y=1.02)
plt.show()
```
